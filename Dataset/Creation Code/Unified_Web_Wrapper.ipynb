```python
def get_site_link():
    #extarcting web page from flipcart
    headers = {
              'dnt': '1',
              'upgrade-insecure-requests': '1',
              'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.60 Safari/537.36',
              'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
              'sec-fetch-site': 'same-origin',
              'sec-fetch-mode': 'navigate',
              'sec-fetch-user': '?1',
              'sec-fetch-dest': 'document',
              'referer':"https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off",
              'accept-language': 'en-GB,en-US;q=0.9,en;q=0.8',
          }

    URL = "https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off"
    product_page = requests.get(URL,headers=headers)
    sp=bs(product_page.content, "html.parser")

    product_sites=sp.find("nav", class_ = "yFHi8N").find_all("a", class_ = "ge-49M")
    page_href=[]

    for i in product_sites:
        page_href.append("https://www.flipkart.com" + i["href"])

    return(page_href)
```

```python
# These headers are to denote the targetting webite that this is not spam.
HEADERS = {
          'dnt': '1',
          'upgrade-insecure-requests': '1',
          'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.60 Safari/537.36',
          'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
          'sec-fetch-site': 'same-origin',
          'sec-fetch-mode': 'navigate',
          'sec-fetch-user': '?1',
          'sec-fetch-dest': 'document',
          'referer':"https://www.nike.com/in/",
          'accept-language': 'en-GB,en-US;q=0.9,en;q=0.8',
      }

#Extracting DOM tree using BeautifulSoup
URL = "https://www.nike.com/in/"
PrdPage = requests.get(URL, headers=HEADERS)
sp=bs(PrdPage.content, "html.parser")

#Extracting main product page links
ProductPage = sp.find("div", class_="pre-desktop-menu-dropdown bg-white p10-sm pt4-sm ncss-container")
Sites = ProductPage.findAll("div", class_="pre-menu-column pt4-sm ta-sm-l va-sm-t ncss-col-sm-2")

#Extracting <a> tags
ATag = []
for i in Sites:   
    ATag.append(i.find("a"))

#Extracting href from <a> tags
PageHref = []
for i in ATag:
  PageHref.append(i["href"])

#Printing links of main product pages
for i in PageHref:
  print(i, end = "\n")
```

#### Extracting Product Page links

```python
# This is the function used to extract and print each product link.
def ScrapProductLink(page_href):
    Cnt = 0
    ProductAnchor = []
    ProductHref = []

    for i in PageHref:
        # These headers are to denote the targetting webite that this is not spam.
        HEADERS = {
              'dnt': '1',
              'upgrade-insecure-requests': '1',
              'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.60 Safari/537.36',
              'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
              'sec-fetch-site': 'same-origin',
              'sec-fetch-mode': 'navigate',
              'sec-fetch-user': '?1',
              'sec-fetch-dest': 'document',
              'referer':i,
              'accept-language': 'en-GB,en-US;q=0.9,en;q=0.8',
        }

        URL = i

        #Extracting each product link from main product page
        EachProductTypePage = requests.get(URL, headers = HEADERS)
        soup = bs(EachProductTypePage.content, "html.parser")
        Result = soup.findAll("div", class_ = "product-card__body")
     
        for i in Result:
            ProductAnchor.append(i.find("a"))
        
        for i in ProductAnchor:
            try:
                ProductHref.append(i["href"])
            except :
                pass

        for i in ProductHref:
                print(" product:%d -> \n"%Cnt,i)
                Cnt += 1
              
    return(ProductHref)
```

```python
ProductLink = ScrapProductLink(PageHref)
```

#### Extracting Product Name

```python
def getNikeName(soup):
    t=None

    try:
        title=soup.find("div",class_="pr2-sm css-1ou6bb2")
        t=title.find("h1").text

    except Exception as e:
        t="NA"
        
    return t
```

#### Extracting Product Price

```python
def getNikePrice(soup):
    price = None
    elm1 = None
    
    try:
        actual_price = soup.find("div", class_ = "product-price css-11s12ax is--current-price css-tpaepq").text
        price = "$Actual price:" + actual_price + "\n\n"
        elm1 = actual_price
    except Exception as e:
        actual_price = e
        pass
   
    return(elm1)
```

#### Extracting Product Image URL

```python
def getNikeImgSrc(soup):
    s = None

    try:
        div = soup.find("picture")
        img = div.find("img")
        src = img["src"]
        s = src

    except Exception as e:
        s = e
        pass
        
    return(s)
```

#### Extracting Product Specification

```python
def getNikeSpecification(soup):
    p_info = None
    p_size = None
    p_det = None

    try:
        try:
            div = soup.findAll("div", class_ = "pi-pdpmainbody")

        except:
            pass

        p_info = div.findAll("p").text
        p_info = "#Product Information :" + p_info + "\n\n"
    
    except:
        p_info = "#Product Information: Unavailable" + "\n\n"
        pass
    
    try:
        try:
            p_det = soup.find("b", class_="headline-5").findAll("li").text

        except:
            pass

        p_det = "#Product Details: " + p_det + "\n"

    except:
        p_det = "#Product Details: unavailable" + "\n"

    try:
        try:
            p_size = soup.find("div",class_="mt2-sm css-12whm6j").findAll("label").text

        except:
            pass

        p_size = "#Product Size: "+p_size+"\n"

    except:
        p_size = "#Product Size: unavailable"+"\n"

    Fspec = p_info + p_det + p_size
    
    return(Fspec)
```

```python
def get_data(pageNo):  
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.60 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",  "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    url = 'https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_'+str(pageNo)+'?ie=UTF8&pg='+str(pageNo)
    #print(url)
    r = requests.get(url, headers=headers)#, proxies=proxies)
    content = r.content
    soup = BeautifulSoup(content)
    #print(soup)

    alls = []
    for d in soup.findAll('div', attrs={'class':'zg-grid-general-faceout'}):
       
        name = d.find('div',attrs={'class':'_p13n-zg-list-grid-desktop_truncationStyles_p13n-sc-css-line-clamp-1__1Fn1y'})
        author = d.find('a', attrs={'class':'a-size-small a-link-child'})
        rating = d.find('span', attrs={'class':'a-icon-alt'})
        price = d.find('span', attrs={'class':'_p13n-zg-list-grid-desktop_price_p13n-sc-price__3mJ9Z'})
        

        all1=[]

        try:
          if name is not None:
            all1.append(name.text)
          else:
             all1.append("unknown-product")

          if author is not None:
            # print(author.text)
              all1.append(author.text)
          elif author is None:
              author = d.find('span', attrs={'class':'a-size-small a-color-base'})
              if author is not None:
                  all1.append(author.text)
              else:    
                  all1.append('0')

          if rating is not None:
              all1.append(rating.text)
          else:
              all1.append('-1')

          if price is not None:
           
              all1.append(price.text)
          else:
              all1.append('0')
          alls.append(all1)    
        except Exception as e:
          print("Error ",e)

    return alls
```

```python
data1  = get_data(1)
data2 = get_data(2)
for x in data2:
  data1.append(x)
df = pd.DataFrame(data1,columns=['Name','Author','Rating', 'Price'])
df.to_csv('/content/drive/MyDrive/ML/UPS_Dataset/amazon_products.csv', index=False, encoding='utf-8')
```
